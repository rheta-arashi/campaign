```python
import numpy as np
import matplotlib.pyplot as plt

# Data
a = np.array([1, 2, 3, 4, 5])
b = np.array([2, 3, 4, 6, 8])
c = np.array([1, 5, 8, 9, 10])

# Create a figure and axis
fig, ax = plt.subplots()

# Plot lines with different format styles
ax.plot(a, c, 'r-', label='Line A-C')
ax.plot(b, c, 'g--', label='Line B-C')

# Set labels and title
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_title('Multiple Lines with Different Styles')

# Add legend
ax.legend()

# Show the plot
plt.show()
```
